```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
phi = np.arange(0,1,0.01)
```

```python
S = (phi*np.log(phi))+((1-phi)*np.log(1-phi))
```

```python
plt.figure()
plt.plot(phi, S, 'ro')
```

```python
xi=2.5
F = (phi*np.log(phi))+((1-phi)*np.log(1-phi)) + (xi*phi*(1-phi))

plt.figure()
plt.plot(phi, F, 'bo')
```
